## Write functions

Write a function that takes in a temperature, moles, and volume and returns a pressure using the ideal gas law.
$$PV = nRT$$

```python
R = 8.314 # J/(mol K)
def pressure(n, T, V):
  P = n*R*T/V
  return P

n = 1 # mols
T = 298 # K
V = 22.4/1000 # Liters

print(f'{pressure(n,T,V)/101325:.2f} atm')
```

## Graph Data
Make a graph of $\sin{x}$, $\cos{2x}$, and $\cos{2x} + \sin{x}$, when $-2\pi < x < 2\pi$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(2*x)
y3 = np.cos(2*x) + np.sin(x)
plt.plot(x,y1, label = 'sin x')
plt.plot(x,y2, label = 'cos(2x)')
plt.plot(x,y3, label = 'cos(2x) + sin(x)')
plt.legend()
plt.show()
```

## Pull it all together
Make a trajectory of a ball that is subject to air resistance, given a $x_0$, $v_0$, and $\theta_0$.

```python
dt = 0.01 # s
t = [0] # s
x0 = 0 # m
y0 = 0 # m
v0 = 10 # m/s
theta = np.pi/4 # rad 45 degrees

vx0 = v0*np.cos(theta)
vy0 = v0*np.sin(theta)
g = 9.81 # m/s^2

resistance = [0,0.1,0.5,1,2,5]

for r in resistance:
  y = y0
  vx = vx0
  vy = vy0
  X = [x0]
  Y = [y0]
  while y >= 0:
    t.append(t[-1] + dt)
    x = X[-1] + vx*dt
    y = Y[-1] + vy*dt
    vx += -vx*r*dt
    vy += -vy*r*dt - g*dt
    X.append(x)
    Y.append(y)

  plt.plot(X,Y, label = f'r = {r}')

plt.legend()
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Ball Trajectory')
plt.show()
```
